```python
import chromadb
chroma_client = chromadb.Client()
```

```python
#Create collection.
collection = chroma_client.create_collection(name="cars2")
```

```python
#Add documents.
collection.add(
    documents = ["acura_cars", "bmw_cars", "ford_cars", "kia_cars", "nissan_cars", "toyota_cars", "car_dataset2024"],
    metadatas = [{"source": "acura cars"}, {"source": "bmw cars"}, {"source": "ford cars"}, {"source": "kia cars"}, {"source": "nissan cars"}, {"source": "toyota cars"}, {"price": "car dataset"}],
    ids = ["id1", "id2", "id3", "id4", "id5", "id6", "id7"]
)
```

```python
#Query the collection.

results = collection.query (
    query_texts=["What company manufactures the Titan?"],
    n_results = 2
)
```

```python
# Nissan manufactured the Titan. The word titan does appear in the Ford document.
results
```

```python
#Include embeddings.
results2 = collection.query (
    query_texts=["What company manufactures convertibles?"],
    n_results = 5,
    include = ["distances", "metadatas", "embeddings", "documents"]
)
```

```python
#The five companies listed manufacture convertibles.
results2
```

```python
#Other query.

results3 = collection.query (
    query_texts=["What car is the safest"],
    n_results = 2
)
```

```python
results3
```

```python
#Filter
results4 = collection.query(
    query_texts = ["Sport Hybrid"],
    n_results = 2,
    where = {"source": "acura cars"}
)
```

```python
results4
```

```python
#Filter

results5 = collection.query(
    query_texts = ["first acura design"],
    where = {"source": "acura cars"},
    n_results = 2,
    where_document = {"$contains": "acura"}
)
```
